```python
import pandas as pd
```

```python
repair = pd.read_csv('repair.csv', sep=',')
vehicle = pd.read_csv('vehicle.csv', sep=',')
buy = pd.read_csv('buy.csv', sep=',')
```

```python
print(repair.columns.values)
print(vehicle.columns.values)
print(buy.columns.values)
```

```python
m = buy.shape[0]
n = repair.shape[0]

for i in range(m):
    vehicle_index = vehicle['vin'].tolist().index(buy['vin'][i])
    print(i, vehicle_index)
    vehicle['sale_price'][vehicle_index] = 1.25*buy['purchase_price'][i]
    for j in range(n):
        if(buy['vin'][i]==repair['vin'][j]):
            vehicle['sale_price'][vehicle_index] = vehicle['sale_price'][vehicle_index] + 1.1*repair['repair_cost'][j]

vehicle.to_csv('new_vehicle.csv', index = False)
```
